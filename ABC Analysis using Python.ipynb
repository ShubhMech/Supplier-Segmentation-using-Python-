```python
data=(suppliers.groupby(['supplier','product']).sum()[['Quantity','Usage Value']])
```

```python
data= pd.DataFrame(data)
```

```python
data
```

```python
data['Usage Value'].sum()
```

```python
data['Fraction Usage Value']=data['Usage Value']/data['Usage Value'].sum()
```

```python
data=data.sort_values('Fraction Usage Value',ascending=False)
```

```python
data['% Usage Value']= data['Fraction Usage Value']*100
```

```python
data.shape
```

```python
sum_=0
list_=[]
for i in range(data.shape[0]):
    print(data['% Usage Value'][i])
    sum_ += data['% Usage Value'][i]
    list_.append(sum_)
# data['Cumulative % Usage Value']= 
```

```python
list_
```

```python
data['Cum Sum (% Usage Val)']=list_
```

```python
data
```

```python
data['% Quantity'] = data["Quantity"]/sum(data['Quantity'])
```

```python
data=data[['Quantity','Usage Value','Fraction Usage Value','% Usage Value','% Quantity','Cum Sum (% Usage Val)']]
```

```python
sum_=0
list_=[]
for i in range(data.shape[0]):
    print(data['% Quantity'][i])
    sum_ += data['% Quantity'][i]
    list_.append(sum_)
# data['Cumulative % Usage Value']= 
```

```python
list_= [i*100 for i in list_]
data['Cum Sum % Quantity']= list_
```

```python
list_
```

```python
data=data[['Quantity','Usage Value','Fraction Usage Value','% Usage Value','% Quantity','Cum Sum % Quantity','Cum Sum (% Usage Val)']]
```
